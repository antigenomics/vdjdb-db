```python
import pandas as pd

import numpy as np
```

```python
data = pd.read_csv('GSE309696_filtered_contig_annotations.csv.gz')
data
```

```python
barcodes_dict = {}
chains = ['TRA', 'TRB']


for barcode in data.barcode.unique():
    
    barcodes_dict[barcode] = {}
    for chain in chains:
        data_i = data[(data.barcode == barcode) & (data.chain == chain)]
        barcodes_dict[barcode][chain] = [
            data_i.v_gene.to_list(),
            data_i.d_gene.to_list(),            
            data_i.j_gene.to_list(),
            data_i.cdr3.to_list(),

        ]
```

```python
peptide_dict = {
    'TC-1': np.nan,
    'TC-2': np.nan,
    'TP-1': 'VEALYLVCG',
    'TP-2': 'VEALYLVCG',
    'TS-1': 'VEALYLVSG',
    'TS-2': 'VEALYLVSG'
}
```

```python
data['antigen.epitope'] = data.origin.map(peptide_dict)

data['antigen.gene'] = 'INS'
data['antigen.species'] = 'HomoSapiens'
data['reference.id'] = 'PMID:41315082'
data['mhc.class'] = 'MHCII'
data['mhc.a'] = 'HLA-DQA1*03:01'
data['mhc.b'] = 'HLA-DQB1*03:02'

data['species'] = 'MusMusculus'


data['method.identification'] = 'tetramer-sort'
data['method.singlecell'] = 'yes'
data['method.sequencing'] = 'amplicon-seq'

data['meta.tissue'] = 'spleen and lymph node'
data['meta.cell.subset'] = 'CD4+'


data['chunk.id'] = list(range(1, len(data)+1))
```

```python
data['cdr3.alpha'] = data.barcode.apply(lambda x: barcodes_dict[x]['TRA'][3])
data['v.alpha'] = data.barcode.apply(lambda x: barcodes_dict[x]['TRA'][0])
data['j.alpha'] = data.barcode.apply(lambda x: barcodes_dict[x]['TRA'][2])


data['cdr3.beta'] = data.barcode.apply(lambda x: barcodes_dict[x]['TRB'][3])
data['v.beta'] = data.barcode.apply(lambda x: barcodes_dict[x]['TRB'][0])
data['d.beta'] = data.barcode.apply(lambda x: barcodes_dict[x]['TRB'][1])
data['j.beta'] = data.barcode.apply(lambda x: barcodes_dict[x]['TRB'][2])

data = data.explode(['cdr3.beta', 'v.beta', 'd.beta', 'j.beta']).explode(['cdr3.alpha', 'v.alpha', 'j.alpha'])


data
```

```python
data.rename(columns = {
    'donor': 'meta.subject.id',
    'barcode': 'meta.clone.id'
}, inplace = True)
```

```python
data = data.drop_duplicates(subset=['barcode', 'cdr3.beta', 'v.beta', 'd.beta', 'j.beta', 'cdr3.alpha', 'v.alpha', 'j.alpha'])

data
```

```python
coulmns = ['chunk.id',
 'cdr3.alpha',
 'v.alpha',
 'j.alpha',
 'cdr3.beta',
 'v.beta',
 'd.beta',
 'j.beta',
 'species',
 'mhc.a',
 'mhc.b',
 'mhc.class',
 'antigen.epitope',
 'antigen.gene',
 'antigen.species',
 'reference.id',
 'method.identification',
 'method.frequency',
 'method.singlecell',
 'method.sequencing',
 'method.verification',
 'meta.study.id',
 'meta.cell.subset',
 'meta.subset.frequency',
 'meta.subject.cohort',
 'meta.subject.id',
 'meta.replica.id',
 'meta.clone.id',
 'meta.epitope.id',
 'meta.tissue',
 'meta.donor.MHC',
 'meta.donor.MHC.method',
 'meta.structure.id',
 'comment']
```

```python
for column in set(coulmns).difference(set(data.columns)):
    data[column] = np.nan

data = data[coulmns]
```

```python
data[data['antigen.epitope'].apply(lambda x: pd.notnull(x))].set_index('chunk.id').to_csv('PMID_41315082.txt', sep='\t')
```

```python
data[data['antigen.epitope'].apply(lambda x: pd.isnull(x))].set_index('chunk.id').to_csv('PMID_41315082_negative.txt', sep='\t')
```
